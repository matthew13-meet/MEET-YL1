The data set I analyzed is about how many incidence in every year melanoma has happend.It came from http://lib.stat.cmu.edu/datasets/visualizing.data.zip 

```python
import csv
```

```python
f = open('melanoma.csv' , 'rb')
reader = csv.reader (f)
rownum = 0
year = []
incidence = []
for row in reader :
    if rownum == 0:
        header = row
    else:
        year.append(row[0])
        incidence.append(row[1])
    
    rownum += 1
year = map(int, year)
incidence = map(float, incidence)
    
f.close()
```

```python
fig = plt.figure(num=None, dpi=70 ,facecolor='green' , edgecolor='green')
cancer = fig.add_subplot(111)
cancer.set_title('melanoma year and incidence')
cancer.set_xlabel('year')
cancer.set_ylabel('incidence')
cancer.plot(year, incidence, color = 'purple')
fig.savefig ('melanoma_year_and_incidence.png')
```

The graph shows that most of the years the incidence if melanoma has increased!
